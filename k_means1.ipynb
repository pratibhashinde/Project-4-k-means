```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
```

```
#random data generation
np.random.seed(0)

x,y=make_blobs(n_samples=5000,n_features=2,centers=[[4,4], [-2, -1], [2, -3], [1, 1]],cluster_std=0.9)
```

```
plt.scatter(x[:,0],x[:,1],c=y,marker='.')
```

```
model=KMeans(init = "k-means++", n_clusters = 4, n_init = 12)
model.fit(x)
```

```
labels=model.labels_
print(labels)
```

```
kmean_centers=model.cluster_centers_
print(kmean_centers)
```

```
cluster_number=4
colors=plt.cm.Spectral(np.linspace(0,1,cluster_number)) #colormap for clusters
fig=plt.figure(figsize=(6,4)) #initialise plot
ax=fig.add_subplot(1,1,1) #create plot
for k,color in zip(range(0,cluster_number),colors):
  members=(labels==k)
  cluster=kmean_centers[k]
  ax.plot(x[members,0],x[members,1],'w',markerfacecolor=color, marker='.')
  ax.plot(cluster[0],cluster[1],'o', markerfacecolor=color,  markeredgecolor='k', markersize=6)

ax.set_title('KMeans')
ax.set_xticks(())
ax.set_yticks(())
plt.show()
```
